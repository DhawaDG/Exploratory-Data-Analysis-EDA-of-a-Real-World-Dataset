### Indexes


There are Three types of datatypes. They are:
1. Default Index : By default, pandas assigns an integer index to each row(starting from 0)
2. Custom Index: You can set a custom index
3. Multilevel Index: Allow you multi levels of indexing(like a tree structure)

```python
#default index
import pandas as pd
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
#before column 'PassengerId' we can see indexing of 0,1,2,3... they are default indexing
```

```python
#custom Index
df.set_index('PassengerId',inplace=True)
df.head()
```

```python
#i have correct column but compiler throwing Spelling mismatch error. lets try on other column
```

```python
df.set_index('Pclass',inplace=True)
df.head()
```

```python
# as you see i have set Pclass as index.
```

```python
#multilevel Index
df.set_index(['Survived', 'Name'],inplace=True)
df.head()
```

## Hexagonal Binning 

1. Instead of plotting each individual points, data points are grouped into hexagonal bins
2. The color intensity of each hexagon repesents the no. of points inside that bins
3. This helps reveal underlying patterns without overcrowding

# advantagees over Scatter plots
1. Clearly shows data density pattern
2. Handles massive datasets efficiently
3. Reveals clusters and outliers that scatterplots would hide

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('titanic.csv')
df.head()
```

```python
#lets try to plot Pclass and Age 
```

```python
#plot hexbins

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

# Hexbin plot with correct x and y values
plt.hexbin(df['Pclass'], df['Age'], gridsize=20, cmap='Blues', mincnt=1)

# Add color bar
plt.colorbar(label='Count in hex bin')

# Labels and title
plt.xlabel("Pclass")
plt.ylabel("Age")
plt.title("Hexagonal Binning: Pclass Vs. Age")

# Save and show plot
plt.savefig("hexagonal_binning_Pclass_vs_Age.png")
plt.show()
```

### Contour Plots

1. Contour plots are like topographic maps but for density
2. The closer the contour lines , the denser the data points
3. The are useful for visualizing cluster and peaks in data distributions

```python
#lets plots Contour plots
import seaborn as sns

plt.figure(figsize=(8,5))
sns.kdeplot(x=df['Pclass'],y=df['Age'],cmap='coolwarm',fill=True,thresh=0.05)

plt.xlabel("Pclass")
plt.ylabel("Age")
plt.title("Contour Plot: Pclass Vs. Age")

plt.savefig("Contour_plot_Pclass_vs_age.png")
plt.show()
```
